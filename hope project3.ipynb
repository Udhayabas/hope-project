```python
import pandas as pd
```

```python
df=pd.read_csv("Inputdata - Sheet1.csv")
df.head()
```

```python
#Converting 'Day' column to datetime format
df['Day'] = pd.to_datetime(df['Day'])

# Sorting DataFrame by 'Day'
df = df.sort_values(by='Day')

# Function to convert date to desired format
def convert_to_custom_date_format(date):
    day_of_week = date.strftime('%A')[:3]
    month = date.strftime('%B')[:3]
    return f"{date.day}{month} ({day_of_week})"

# Applying the function to 'Day' column
df['Day'] = df['Day'].apply(convert_to_custom_date_format)



# Displaying the DataFrame
print(df)
```

```python
df=pd.DataFrame(df)
df.fillna(0, inplace=True)
df=df.reset_index(drop=True)
df.head()
```

```python
data = df[['Day', 'Campaign name', 'Amount spent (INR)', 'Results', 'CTR (link click-through rate)', 'Link clicks', 'Landing page views']]

# Rename columns
data.rename(columns={'Day':'Date','Amount spent (INR)': 'AD Spent', 'Results': 'No of leads', 'CTR (link click-through rate)': 'CTR', 'Landing page views': 'LP Views'}, inplace=True)

data.fillna(0, inplace=True)

# Calculate Cost per Lead
data['Cost per Lead'] = data['AD Spent'] / data['No of leads']

# Calculate Member passed
data['Member passed'] = data['LP Views'] / data['Link clicks']*100
  

# Calculate LP Conversion
data['LP Conversion%'] = data['No of leads'] / data['LP Views']*100


data.head()
```

```python
data = data[['Date', 'Campaign name', 'AD Spent', 'No of leads','Cost per Lead', 'CTR',
       'Link clicks', 'LP Views', 'Member passed',
       'LP Conversion%']]
data.head()
```
